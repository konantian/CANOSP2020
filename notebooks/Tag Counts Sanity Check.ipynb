```python
import pandas as pd
```

```python
df = pd.read_csv("../data/preprocessed_class_input_intersection.csv")
df
```

```python
with open("../candidate_tags/all_methods_intersection", "r") as f:
    candidate_tags = [line.strip() for line in f.readlines()]
    
candidate_tags
```

```python
tag_counts = []

for tag in candidate_tags:
    tag_counts.append((tag, df[tag].sum()))

tag_counts.sort(key=lambda x: x[1], reverse=True)

tag_counts, len(tag_counts)
```
