```python
"""
To do: create a json that has ticket id and tags assigned to it.
        Do it for both TF-IDF and YAKE!
       do analysis with 1 word
"""

import yake
import json
import re
import operator
```

```python
# Get tickets we want to tag

with open("Data/sample_data.json") as f:
  data = json.load(f)

tickets = {}
tickets = data['tickets']

with open('data/ticket_data.json', 'w') as json_file:
  json.dump(tickets, json_file)
```

```python
def get_keywords(text):
    
    language = "en"
    max_ngram_size = 2
    deduplication_thresold = 0.9
    deduplication_algo = 'seqm'
    windowSize = 1
    numOfKeywords = 20

    custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
    keywords = custom_kw_extractor.extract_keywords(text)
    
    return keywords
```

```python
# later change it to Michael's code

def preprocessing(text):
    # lowercase
    text=text.lower()
    
    #remove tags
    text=re.sub("</?.*?>"," <> ",text)
    
    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)
    
    return text
```

```python
"""
for each ticket in the json file:
    - Combine title and content
    - Apply preprocessing
    - Get tags
    - Add to the dict
"""
tags_dict = {}

for i in range(30000):
    text = tickets[i]['title'] + ' ' + tickets[i]['content']
    text = preprocessing(text)
    
    #print(text)
    try:
        tags_list = get_keywords(text)
    except:
        print('An error occurred.')
        continue
    #print(tags_list)
    
    for tag_value_pair in tags_list:
        tag = tag_value_pair[0]
        value = tag_value_pair[1]
        
        if value <= 0.02:
            if tag in tags_dict:
                tags_dict[tag] += 1
            else:
                tags_dict[tag] = 1
                
with open('YAKE_ALL_tags.txt', 'w') as file:
    file.write(json.dumps(tags_dict))
    

usefull_tag_dict = {}

for key, value in tags_dict.items():
    if value > 10:
        usefull_tag_dict[key] = value

print(len(usefull_tag_dict))
        
with open('YAKE_tags.txt', 'w') as file:
    file.write(json.dumps(usefull_tag_dict))
    
```

```python
sorted_list = sorted(usefull_tag_dict.items(), key=operator.itemgetter(1), reverse = True)

print("{:<35} {:<7}".format('tag', 'count'))

for k, v in sorted_list:
    print("{:<35} {:<7}".format(k, v))
```
